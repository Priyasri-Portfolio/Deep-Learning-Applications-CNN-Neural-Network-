```python
#Question 1
# Import libraries
import numpy as np
import tensorflow as tf
import random
from keras.datasets import mnist
from keras.models import Sequential
from keras.layers import Dense, Flatten, Dropout, Conv2D, MaxPooling2D
from keras.utils import to_categorical
from keras.optimizers import RMSprop
```

```python
#load the data
(x_train, y_train), (x_test, y_test) = mnist.load_data()
```

```python
# Normalize the pixel values(0-255 in to 0/1)
x_train = x_train.astype('float32') / 255
x_test = x_test.astype('float32') / 255

#Converting labels to One-hot encoding(digits 0-9  -->binary matrix of size 10)
y_train = to_categorical(y_train, 10)
y_test = to_categorical(y_test, 10)

# Reshape input data to include channel dimension (grayscale вҶ’ 1 channel)
x_train = x_train.reshape(x_train.shape[0], 28, 28, 1)
x_test = x_test.reshape(x_test.shape[0], 28, 28, 1)

# Set seeds for reproducibility
np.random.seed(123)
tf.random.set_seed(123)

```

```python
 #Build CNN model
# -----------------------------
model = Sequential([
    # (i) First convolution layer with 32 filters of size 3x3, ReLU activation
    Conv2D(32, (3, 3), activation='relu', input_shape=(28, 28, 1)),
    # Max pooling over 2x2 blocks
    MaxPooling2D((2, 2)),

    # (ii) Second convolution layer with 64 filters of size 3x3, ReLU activation
    Conv2D(64, (3, 3), activation='relu'),
    # Max pooling over 2x2 blocks
    MaxPooling2D((2, 2)),

    # (iii) Third convolution layer with 64 filters of size 3x3, ReLU activation
    # Output shape will be (3, 3, 64)
    Conv2D(64, (3, 3), activation='relu'),
    # Flatten the output before dense layers
    Flatten(),

    # (iv) Dense layer with 64 units, ReLU activation
    Dense(64, activation='relu'),

    # (v) Output layer with 10 units (digits 0вҖ“9), softmax activation
    Dense(10, activation='softmax')
])
```

```python
# Compile the model
# RMSprop optimizer with learning rate 0.01
model.compile(optimizer=RMSprop(learning_rate=0.01),
              loss='categorical_crossentropy',
              metrics=['accuracy'])

# Print model summary to verify architecture
model.summary()
```

```python
# -----------------------------
# Train the model
# -----------------------------
# 5 epochs, minibatch size = 64, validation split = 10%
history = model.fit(x_train, y_train,
                    batch_size=64,
                    epochs=5,
                    validation_split=0.1,
                    verbose=1)

# -----------------------------
# Evaluate on test data
# -----------------------------
test_loss, test_acc = model.evaluate(x_test, y_test, verbose=1)
print(f'\nTest accuracy: {test_acc * 100:.6f}%')
```

```python
from ISLP.torch import SimpleModule, SimpleDataModule, ErrorTracker
# Load IMDB dataset
max_num_workers = 10

#wrap data into a DataModule (handles batching, validation split )
(imdb_train, imdb_test) = load_tensor(root='data/IMDB')
imdb_dm = SimpleDataModule(
    imdb_train,
    imdb_test,
    validation=2000,# reserving 2000 samples for validation
    num_workers=min(6, max_num_workers),#number of parallel workers for loading
    batch_size=512 # batch size for training
)
```

```python
# Defining the model
class IMDBModel(nn.Module):
    def __init__(self, input_size, hidden_units=16, dropout_rate=0.0):
        super(IMDBModel, self).__init__()
        #build a simple feedforward network with optional dropout
        self.network = nn.Sequential(
            nn.Linear(input_size, hidden_units), # first hidden layer
            nn.ReLU(),                           #activation function
            nn.Dropout(p=dropout_rate),          #dropout for regularization
            nn.Linear(hidden_units, hidden_units), # second hidden layer
            nn.ReLU(),                              #activation function
            nn.Dropout(p=dropout_rate),             #dropout again
            nn.Linear(hidden_units, 1)              #output layer(binary classification)
        )

    def forward(self, x):
        return torch.flatten(self.network(x))

# Training + evaluation function
def train_and_evaluate_model(hidden_units, dropout_rate):
  # create model with given hidden units and dropout
    imdb_model = IMDBModel(
        input_size=imdb_test.tensors[0].size()[1], # number of input features
        hidden_units=hidden_units,
        dropout_rate=dropout_rate
    )
    #define optimizer (RMSprop with learning rate 0.01)
    optimizer = RMSprop(imdb_model.parameters(), lr=0.01)

    #wrap model in ISLP SimpleModuel for binary classification
    imdb_module = SimpleModule.binary_classification(imdb_model, optimizer=optimizer)

    #logger to save trainng progress
    imdb_logger = CSVLogger("logs", name=f"IMDB_units_{hidden_units}_dropout_{dropout_rate}")

    #Trainer handles training loop(30 epochs, deterministic for reproducibility)
    trainer = Trainer(
        deterministic=True,
        max_epochs=30,
        logger=imdb_logger,
        callbacks=[ErrorTracker()] # track errors during training
    )

    #train the model
    trainer.fit(imdb_module, datamodule=imdb_dm)

    #Evaluate on the test set
    test_res = trainer.test(imdb_module, datamodule=imdb_dm)

    #print results for this configuration
    print(f"Configuration: {hidden_units} units, {dropout_rate} dropout", test_res)
    return test_res

# Run all configurations
results = []

#loop over different hidden unit sizes and dropout settings
for units in [16, 32, 64]:
    for dropout in [0.0, 0.3]:
        results.append(train_and_evaluate_model(units, dropout))
```
